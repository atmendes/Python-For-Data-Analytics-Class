```python
#Create a variable for Kids from 0-1
```

```python
kids=['[0-10)']
```

```python
#filter by kids 0-10
```

```python
School_age_Children = df[df.age.isin(kids)]
```

```python
#Check count of rows for school children
```

```python
School_age_Children.count()
```

```python
#group by age to find out the oldest age group to compare with the youngest age group
```

```python
df.groupby(['age'])['patient_nbr'].count()
```

```python
#Create variable for Oldest age group
```

```python
Oldest=['[90-100)']
```

```python
Oldest_Group = df[df.age.isin(Oldest)]
```

```python
#Check the count of the oldest age group- much more than the youngest as to be expected
```

```python
Oldest_Group.count()
```

```python
#Check the race breakdown for school age group
```

```python
School_age_Children.groupby(['race'])['patient_nbr'].count()
```

```python
#Check Time in hospital for School age group
```

```python
School_age_Children.groupby(['race'])['time'].count()
```

```python
#Check time in hospital for oldest group, based on race
```

```python
Oldest_Group.groupby(['race'])['time_in_hospital'].mean()
```

```python
School_age_Children.groupby(['race'])['time_in_hospital'].mean()
```

```python
#COnclusion for first section- School age children stayed in the hospital for a shorter amount of time compared with the oldest group. For the oldest group, Hispanic people stayed in the hosital the longest, whereas in 
#the yonger group, it was African Americans (beside other)
```
